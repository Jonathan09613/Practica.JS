7. Aplica el formato a los títulos de columnas el formato correcto

Isin y ~ Isin

```python
lista_mapas = ['de_dust2', 'de_mirage', 'de_inferno']
df_lista = df_CS[df_CS['map'].isin(lista_mapas)][['map', 'time_left', 'ct_score', 't_score']]
print(df_lista)
```

```python
mapas_a_excluir = ['de_dust2', 'de_mirage']
df_resultado = df_CS[~df_CS['map'].isin(mapas_a_excluir)][['map', 'time_left', 'ct_score', 't_score']]
print(df_resultado)
```

```python
#Ej.1
def round_phase(time_left):

    if time_left >= 120:
      return 'alto'
    elif time_left >= 60:
      return 'medio'
    elif time_left <= 60:
      return 'bajo'
    else:
        return 'unknown'
```

```python
df_CS['round_phase'] = df_CS['time_left'].apply(round_phase)
print(df_CS['round_phase'])
```
